```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
notebook_path = os.path.abspath(os.path.join(os.getcwd(), '../notebooks'))
input_file_path = rf'{notebook_path}\data\processed'
```

Steps

Reads the CSV file and parses datetime.

Lagged Features: Adds past prices (lag_1, lag_2, etc.).

The code preprocesses and engineers time series features, trains a neural network using time-series cross-validation, and evaluates the model's performance with comprehensive metrics.
```python
# Step 1: Load the Data
year = "2023"
ccy = "gbpusd"
file_path = rf'{input_file_path}/{year}_{ccy}_tick_prev_price_as_base.csv'

# Load the dataset
df = pd.read_csv(file_path)
df["datetime"] = pd.to_datetime(df["datetime"])

# Step 2: Feature Engineering

# Create lagged features
lags = [1, 2, 3, 5, 10]
for lag in lags:
    df[f'lag_{lag}'] = df['price'].shift(lag)

# Rolling mean and standard deviation
rolling_windows = [5, 10, 20]
for window in rolling_windows:
    df[f'rolling_mean_{window}'] = df['price'].rolling(window=window).mean()
    df[f'rolling_std_{window}'] = df['price'].rolling(window=window).std()

# Exponential moving averages (EMA)
ema_spans = [5, 10, 20]
for span in ema_spans:
    df[f'ema_{span}'] = df['price'].ewm(span=span, adjust=False).mean()

# Momentum indicators
df['roc'] = (df['price'] - df['price'].shift(5)) / df['price'].shift(5) * 100  # Rate of Change
df['rsi'] = 100 - (100 / (1 + (df['price'].diff().clip(lower=0).rolling(14).mean() /
                               df['price'].diff().clip(upper=0).abs().rolling(14).mean())))

# Drop missing values caused by rolling/lagging
df.dropna(inplace=True)

# Define features and target
features = [col for col in df.columns if col.startswith(('lag_', 'rolling_', 'ema_', 'roc', 'rsi'))]
df['target'] = (df['directions'] > 0).astype(int)  # Binary target: 1 for up, 0 for down
X = df[features]
y = df['target']

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: TimeSeriesSplit for Train-Test Splitting
tscv = TimeSeriesSplit(n_splits=5)

# Prepare for storing results
accuracies = []
conf_matrices = []
classification_reports = []

# Step 4: Train and Evaluate Model
for train_index, test_index in tscv.split(X_scaled):
    # Train-test split
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Define Neural Network Model
    model = Sequential([
        Dense(64, activation='relu', input_shape=(X_train.shape[1],), kernel_regularizer=tf.keras.regularizers.l2(0.01)),
        Dropout(0.3),
        Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
        Dropout(0.3),
        Dense(1, activation='sigmoid')  # Sigmoid for binary classification
    ])

    # Compile the Model
    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # Train the Model
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)

    # Evaluate the Model
    y_pred_probs = model.predict(X_test)
    y_pred = (y_pred_probs > 0.5).astype(int).flatten()  # Convert probabilities to binary labels

    # Collect Metrics
    accuracy = accuracy_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    classification_report_text = classification_report(y_test, y_pred)

    accuracies.append(accuracy)
    conf_matrices.append(conf_matrix)
    classification_reports.append(classification_report_text)

    print(f"Fold Accuracy: {accuracy}")
    print(f"Confusion Matrix:\n{conf_matrix}")
    print(f"Classification Report:\n{classification_report_text}")

# Step 5: Display Average Results
avg_accuracy = np.mean(accuracies)
print(f"\nAverage Accuracy Across Folds: {avg_accuracy}")

# Step 6: Plot Training History (from last fold)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Training History (Last Fold)')
plt.show()
```

What Are Folds?
In the context of cross-validation, a fold refers to one partition or split of the dataset used during the training and testing process. Cross-validation is a technique to assess the performance of a model by dividing the dataset into multiple segments (folds) and training/testing the model on different combinations of these segments.

For example, if you perform 5-fold cross-validation:

The dataset is split into 5 equal parts (folds).
The model is trained on 4 of these folds and tested on the remaining 1 fold.
This process is repeated 5 times, with each fold serving as the test set once.
The average performance across all folds gives a robust estimate of how the model will perform on unseen data.

```python
def backtest_strategy(df, y_pred, transaction_cost=0):
    """
    Perform backtesting on a dataset based on model predictions.
    
    Parameters:
        df (pd.DataFrame): DataFrame containing price data and other features.
        y_pred (np.array or list): Model predictions (0 or 1) for the target.
        transaction_cost (float): Cost per trade as a fraction of price (default is 0).

    Returns:
        dict: Backtesting results including metrics and cumulative returns.
    """
    df = df.copy()  # Avoid modifying the original DataFrame

    # Step 1: Add Predictions to DataFrame
    df['predicted_signal'] = 0  # Initialize column
    df.loc[df.index[-len(y_pred):], 'predicted_signal'] = y_pred  # Add predictions to the test period

    # Step 2: Generate Trading Signals
    # Buy (1) if predicted_signal == 1, Sell (-1) if predicted_signal == 0
    df['position'] = df['predicted_signal'].apply(lambda x: 1 if x == 1 else -1)

    # Step 3: Calculate Returns
    df['price_change'] = df['price'].diff()  # Price change
    df['strategy_return'] = df['position'].shift(1) * df['price_change']  # Strategy returns
    if transaction_cost > 0:
        # Subtract transaction costs
        trades = df['position'].diff().abs()  # Count trades (1 for every position change)
        df['strategy_return'] -= trades * transaction_cost

    # Step 4: Calculate Cumulative Returns
    df['cumulative_strategy_return'] = df['strategy_return'].cumsum()
    df['cumulative_market_return'] = df['price_change'].cumsum()

    # Step 5: Calculate Metrics
    total_return = df['strategy_return'].sum()
    win_rate = (df['strategy_return'] > 0).mean()
    sharpe_ratio = df['strategy_return'].mean() / (df['strategy_return'].std() + 1e-10)  # Avoid division by zero
    drawdown = (df['cumulative_strategy_return'].cummax() - df['cumulative_strategy_return']).max()

    # Step 6: Plot Results
    plt.figure(figsize=(10, 6))
    plt.plot(df['cumulative_strategy_return'], label='Strategy Return')
    plt.plot(df['cumulative_market_return'], label='Market Return')
    plt.xlabel('Time')
    plt.ylabel('Cumulative Returns')
    plt.legend()
    plt.title('Backtesting Results')
    plt.show()

    # Return Results
    results = {
        "Total Return": total_return,
        "Win Rate": win_rate,
        "Sharpe Ratio": sharpe_ratio,
        "Max Drawdown": drawdown
    }
    return results
```

```python
# Assuming `df` is your DataFrame and `y_pred` is the model's predictions
results = backtest_strategy(df, y_pred, transaction_cost=0.001)  # Adjust transaction_cost as needed

# Print Backtesting Metrics
print(f"Total Return: {results['Total Return']}")
print(f"Win Rate: {results['Win Rate']:.2%}")
print(f"Sharpe Ratio: {results['Sharpe Ratio']:.2f}")
print(f"Max Drawdown: {results['Max Drawdown']:.2f}")
```

Backtesting

Parameters:

initial_balance: The starting capital for the backtest.
lot_size: The size of each trade (in units of the currency).
Logic:

A trade is placed based on the model's prediction:
If the model predicts "up" (1), we simulate a buy.
If the model predicts "down" (0), we simulate a sell.
Profit or loss is calculated based on whether the model's prediction aligns with the actual movement (y_test).
Trade Profit Calculation:

price * 0.0001 * lot_size is a simple approximation for pip movement profit/loss.
This assumes 1 pip = 0.0001 for most currency pairs (adjust as needed).
Outputs:

balance_history: The account balance over time.
trade_log: Detailed log of each trade, including tick, price, predicted and actual direction, profit, and updated balance.
